```python
#Lets do some data cleaning by changing all column names to small letters and using _ to join words together
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

```python
#create a new column name 'discount_price'
```

```python
df['discount_price'] = df['list_price']*df['discount_percent']*.01
```

```python
df.head(5)
```

```python
#create a new column 'sale_price' that shows the actual sales price
```

```python
df['sale_price'] = df['list_price']-df['discount_price']
```

```python
df.head(5)
```

```python
#create a new column to show the profit made after sale
```

```python
df['profit'] = df['sale_price']- df['cost_price']
```

```python
df.head()
```

```python
#let's see the data types of each column
df.dtypes
```

```python
#convert the date format to date-time
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format = "%Y-%m-%d")
```

```python
df.head(5)
```

```python
#drop the columns that are irrelevant to analysis
```

```python
df.drop(columns =['cost_price', 'list_price', 'discount_percent'], inplace = True)
```
